### LLMs

#### What are LLMs?


image.png

The true ‘magic’ of LLMs lies in the parameters, which are crafted during the intensive process of model training. Training involves condensing a vast amount of internet text data. For instance, llama-2-70b utilized around **10TB** of web-crawled text and required significant computational resources:

- Compute Power: Utilized about 6,000 GPUs over 12 days.
- Cost: Approximated at $2M.
- Compute Operations: Reached around 1e24 FLOPS (floating-point operations per second).

#### Training of an LLM
**Stage One: Pretraining**

This stage creates what we call the base model, essentially a sophisticated document generator.

**Stage Two: Fine Tuning**

Fine tuning, or ‘alignment,’ transforms the base model into an assistant. This involves training on high-quality, manually generated Q&A sets, typically comprising around 100,000 conversations.

#### So does this means simple one word is called a token in LLMs?

here is a way to estimate from OpenAI:

Tokens are the building blocks of text that OpenAI models process. They can be as short as a single character or as long as a full word, depending on the language and context. Spaces, punctuation, and partial words all contribute to token counts. This is how the API internally segments your text before generating a response.

Helpful rules of thumb for English:

1 token ≈ 4 characters

1 token ≈ ¾ of a word

100 tokens ≈ 75 words

#### Disadvantages of close-source LLMs





*   Privacy Risk: Data leaves the internal system and is sent to external servers.
*   Cost: Ongoing fees for cloud services or APl usage.
*   Limited Customizability: Less control and limited ability to modify.
*   Dependency on internet Connection: Does not function without internet access.
*   Security Concerns: Risk of data being accessed by third parties.
*   Flexibility: Limited integration and adaptation to specific applications and workflow
*   Dependence: Long-term dependence on external service providers.
*   Transparency: No insight into the code and functionality of the model.
*   Risk: Potentially biased responses that are difficult to verify and correct.

#### What if I don't have good enough hardware? (Quantization)

Basic ldea: lnstead of storing and processing the numbers in a model with high precision (e.g.32-bit), they are reduced to a lower precision (e.g., 8-bit or 4-bit).

Example: A number that is normally stored as a 32-bit floating point could be stored as an 8-bior 4-bit integer.

Advantages:

*   Save Memory Space: Fewer bits mean less memory is required.
*   Faster Calculations: Smaller number formats can be processed more quickly, especially onspecialized hardware.
*   Q8: Here, the numbers in the model are reduced to 8-bit precision. This is a moderate reductiand offers a good compromise between memory space, computational performance, and modaccuracy.
*   Q4: Here, the numbers in the model are reduced to 4-bit precision. This saves even morememory space and computational power but can lead to a greater loss of accuracy.

For example - [Llama 3 8B Q4 model](https://huggingface.co/spiharsh/Meta-Llama-3-8B-Q4/tree/main) is less than 5GB.

#### LM Studio

#### Structured Propting (Covered by Dr. Lance)


A structured prompt consists of a (Modifier) + (Topic) + (Multiple Modifiers).

These elements together form the optimized prompt.

*   Modifier: Specifies the type of desired response. Examples: "Twitter thread,"blog post," "research paper," etc.

*   Topic: The main subjeet or question to be addressed in the response.

*   Additional Modifiers: Additional specific requirements or details to beconsidered in the response, Examples: Target audience, keywords used, style, length, structure, etc.

Example: Write a **blog post** about **healthy eating**. Address it to (working professionals) and use keywords that are relevant for *search engine optimization*. Write the text in a **simple, understandable style** so that it is easy to read and comprehend, Thelength should be **800 words**, and the text should be well-structured.

1) In disaster communication, it should be a good idea to tell your chatbot "if you are not confident in providing suggestions, you can acknowledge that the question is beyond you can answer with confidence, then, either 1) tell the user your best guess if you have a relatively confident suggestions, or 2) tell the user you are not designed to answer this question.
